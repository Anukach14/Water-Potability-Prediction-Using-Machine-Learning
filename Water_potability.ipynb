```python
from sklearn.preprocessing import LabelEncoder

labelencoder_Y = LabelEncoder()
labelencoder_Y.fit_transform(df.iloc[:,1].values)
df.iloc[:,1] = labelencoder_Y.fit_transform(df.iloc[:,1].values)
df.iloc[:,1] #Categorical data values
```

```python

from sklearn.model_selection import train_test_split

X = df.iloc[:,2:7].values
Y = df.iloc[:,1].values
X_train, X_test, Y_train, Y_test = train_test_split( X, Y ,test_size = 0.20, random_state = 0) #Splitting and Scaling of dataset
```

```python

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print("X_train shape : ",X_train.shape)
print("X_test shape : ",X_test.shape)
print("Y_train shape : ",Y_train.shape)
print("Y_test shape : ",Y_test.shape) #Dimensions of dataset
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state = 40)
logreg.fit(X_train, Y_train)
accuracy_logreg = logreg.score(X_train, Y_train) #Using Logistic Regression
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski' , p=2)
knn_model.fit(X_train , Y_train)
accuracy_knn = knn_model.score(X_train, Y_train) #Using KNN
```

```python
from sklearn.svm import SVC

svm_model = SVC( kernel = 'linear',random_state = 42)
svm_model.fit(X_train , Y_train)
accuracy_svm = svm_model.score(X_train, Y_train) #Using Support Vector Classifier
```

```python

from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()
nb_model.fit(X_train , Y_train)
accuracy_nb = nb_model.score(X_train, Y_train) #Using Naive Bayes
```

```python

from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier( criterion = 'entropy' , random_state = 1)
dt_model.fit(X_train , Y_train)
accuracy_dt = dt_model.score(X_train, Y_train) #Using Decision Tree
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators = 6 ,criterion = 'entropy' , random_state = 25)
rf_model.fit(X_train , Y_train)
accuracy_rf = rf_model.score(X_train, Y_train) #Using Random Forest
```

```python
print("Logistic Regression Score : ",accuracy_logreg)
print("KNN Score : ",accuracy_knn)
print("Support Vector Classifier Score : ",accuracy_svm)
print("Naive bayes Score : ",accuracy_nb)
print("Decision Tree Score : ",accuracy_dt)
print("Random Forest Score : ",accuracy_rf) #Accuracy
```
